```python
## Plot the range ramp igven a constant translation

import numpy as np
import matplotlib.pyplot as plt
font_size=16
plt.rcParams.update({'font.size': font_size})

# distance
x = np.arange(100)

# height from platform
h = 50

# uniform horizontal translation
vh = 1

# incidence angle from asc, dsc
theta_a = np.arctan(x/h)
theta_d = np.arctan((np.max(x)-x)/h)

# LOS motion from asc, dsc
vlos_a = -vh*np.sin(theta_a)
vlos_d = +vh*np.sin(theta_d)


plt.figure(figsize=[6,4])
plt.plot(x, vlos_a, c='r', lw=3, label='ASC')
plt.plot(x, vlos_d, c='b', lw=3, label='DSC')
plt.xlabel('Distance')
plt.ylabel('Observed motion')
plt.legend(loc='upper right')
plt.show()
```
